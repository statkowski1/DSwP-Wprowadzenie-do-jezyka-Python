```python
#Zadanie 1

from timeit import timeit
import random

setup = """
from array import array
import random
"""
stmt1 = """
tab_of_chars = array('u', [chr(random.randint(32, 126)) for _ in range(1_000)])
tab_of_ints = array('i', [random.randint(1, 100) for _ in range(1_000)])
tab_of_longs = array('l', [random.randint(1, 100) for _ in range(1_000)])
"""
stmt2 = """
list_of_chars = [chr(random.randint(32, 126)) for _ in range(1_000)]
list_of_ints = [random.randint(1, 100) for _ in range(1_000)]
list_of_longs = [random.randint(1, 100) for _ in range(1_000)]
"""

print("Czas inicjowania tablicy znaków:")
print(timeit(stmt1, setup, number=100))
print("Czas inicjowania listy znaków:")
print(timeit(stmt2, setup, number=100))

print("Czas inicjowania tablicy int:")
print(timeit(stmt1, setup, number=100))
print("Czas inicjowania listy int:")
print(timeit(stmt2, setup, number=100))

print("Czas inicjowania tablicy long:")
print(timeit(stmt1, setup, number=100))
print("Czas inicjowania listy long:")
print(timeit(stmt2, setup, number=100))
```

```python
#Zadanie 2

import datetime
from array import array
import random

start_time = datetime.datetime.now()

tab_of_floats = array('f', [random.random() for _ in range(1_000)])
with open('floats_array.bin', 'wb') as file_arr:
    tab_of_floats.tofile(file_arr)

tab_of_floats_loaded = array('f')
with open('floats_array.bin', 'rb') as file_arr:
    tab_of_floats_loaded.fromfile(file_arr, 1_000)

end_time = datetime.datetime.now()
time_array = end_time - start_time
print("Czas operacji zapisu i odczytu dla tablicy:", time_array)

start_time = datetime.datetime.now()

list_of_floats = [random.random() for _ in range(1_000)]
with open('floats_list.txt', 'w') as file_arr:
    file_arr.writelines('\n'.join([str(x) for x in list_of_floats]))

list_of_floats_loaded = []
with open('floats_list.txt', 'r') as file_list:
    list_of_floats_loaded = file_list.readlines()
list_of_floats_loaded = [float(x.strip()) for x in list_of_floats_loaded]

end_time = datetime.datetime.now()
time_list = end_time - start_time
print("Czas operacji zapisu i odczytu dla listy:", time_list)

#Wnioski:
#Operacje zapisu i odczytu danych dla tablicy są szybsze niż dla listy (tablice są bardziej zoptymalizowane pod
#kątem operacji zapisu i odczytu danych)
```

```python
#Zadanie 3

from collections import deque
import timeit

deque_time_append = timeit.timeit("deque_obj.append(1)", setup="from collections import deque; deque_obj = deque()", number=10000)
deque_time_appendleft = timeit.timeit("deque_obj.appendleft(1)", setup="from collections import deque; deque_obj = deque()", number=10000)
print("Czas operacji append na deque:", deque_time_append)
print("Czas operacji appendleft na deque:", deque_time_appendleft)

list_time_append = timeit.timeit("list_obj.append(1)", setup="list_obj = []", number=10000)
list_time_insert = timeit.timeit("list_obj.insert(0, 1)", setup="list_obj = []", number=10000)
print("Czas operacji append na liście:", list_time_append)
print("Czas operacji insert(0) na liście:", list_time_insert)
```

```python
#Zadanie 4

from collections import namedtuple

with open('zamowienia.csv', 'r', newline='', encoding='cp1252') as file:
    labels = file.readline().strip().split(';')
    labels = [label.replace(' ', '_') for label in labels]
    Order = namedtuple('Order', labels)
    print("Etykiety krotek:", labels)
    
    for line in file:
        data = line.strip().split(';')
        order = Order(*data)
        print(order)
```

```python
#Zadanie 5

import random

def top_10_percent(arr):
    arr.sort(reverse=True)
    return arr[:int(len(arr)*0.1)]

values = [random.randint(1, 1000) for _ in range(100)]
print('Wylosowane wartości: ', values)
print('Posortowane wylosowane wartości malejąco: ', sorted(values, reverse=True))
print('10% najwyższych wartości:', top_10_percent(values))
```

```python
#Zadanie 6

from collections import Counter, deque

def create_kolo_fortuny(*args):
    counter = Counter(args)
    return deque(counter.elements())

wheel = create_kolo_fortuny('A', 'B', 'C', 'A', 'B', 'A')
print(wheel)
```

```python
#Zadanie 7

import random
import time
import numpy as np
from collections import deque

def spinit(*args):
    wheel_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
    tmp = 0
    waits = np.logspace(0.0, 1.0, num=len(wheel_list)) / len(wheel_list)
    waits2 = np.logspace(0.0, 1.1, num=len(wheel_list)+1) / (len(wheel_list)+1)
    for arg in args:
        index = wheel_list.index(arg)
        if arg == 'A':
            for i, tick in enumerate(wheel_list+list(wheel_list[0])):
                print(f'{tick}', end='')
                time.sleep(waits2[index])
                print('\r', end='')
                if tick == arg and tmp > 0:
                    break
                tmp += 1
        else:
            for i, tick in enumerate(wheel_list):
                print(f'{tick}', end='')
                time.sleep(waits[index])
                print('\r', end='')
                if tick == arg and tmp > 0:
                    break
                tmp += 1
        print('\n')
        tmp = 0

spinit('A', 'G', 'J', 'F')
```
